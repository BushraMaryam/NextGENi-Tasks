```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
```

```python
#for paklap
site1_url="https://www.paklap.pk/dell-inspiron-15-3511-11th-gen-core-i7-laptop-price-reviews-pakistan.html"
response_1=requests.get(site1_url)
content_1=response_1.content
parsed_content_1 = BeautifulSoup(content_1, 'html.parser')
print(parsed_content_1)
```

```python
#specifications
table_1=parsed_content_1.find("table",class_="data table additional-attributes")
labels_1=[]
values_1=[]
for row in table_1.find_all('tr'):
   
    for cell in row.find_all('th'):
        labels_1.append(cell.get_text())
    for cell in row.find_all('td'):
            values_1.append(cell.get_text())
datalist_1={'labels':labels_1,'values':values_1}

```

```python
#Price
price_1=parsed_content_1.find("span",class_="price")
labels_1.append("Price")
values_1.append(price_1.get_text())
df1=pd.DataFrame(datalist_1)
df1
```

```python
#cleaning
#Removing Duplicate Rows
duplicates=df1[df1.duplicated()]
print('Duplicate Rows \n',duplicates)
#Removing nul value rows
df1.replace('-',pd.NA,inplace=True)
missing_values=df1.isnull().sum()
print("Missing Values \n",missing_values)

```

```python
df1_cleaned=df1.dropna()
df1_cleaned=df1_cleaned.drop_duplicates()
df1_cleaned.reset_index(drop=True,inplace=True)
df1_cleaned
```
